```julia
@show space(env_arr[1].ul)
@show space(env_arr[1].l)
```

```julia
function tn_mul_4(X, env_arr)
    @tensor Y[-1, -2; -3] := env_arr.u[-1, 1, 2] * env_arr.d[-2, 3, 1] *
                             X[2, 3, -3]
    return Y
end


V_Z2_even = Z2Space(0 => 1, 1 => 0) 
X = TensorMap(randn,  space(env_arr[1].u, 1) ⊗ space(env_arr[1].d, 1),  V_Z2_even)
spectrum_even, _ = eigsolve(X, 12, :LM) do x return tn_mul_4(x, env_arr[1]) end

V_Z2_odd = Z2Space(0 => 0, 1 => 1) 
X = TensorMap(randn,  space(env_arr[1].u, 1) ⊗ space(env_arr[1].d, 1),  V_Z2_odd)
spectrum_odd, _ = eigsolve(X, 12, :LM) do x return tn_mul_4(x, env_arr[1]) end

V_Z2 = Z2Space(0 => 1, 1 => 1) 
X = TensorMap(randn,  space(env_arr[1].u, 1) ⊗ space(env_arr[1].d, 1),  V_Z2)
spectrum_all, _ = eigsolve(X, 13, :LM) do x return tn_mul_4(x, env_arr[1]) end

@show spectrum_even
@show spectrum_odd
@show spectrum_all

nothing
```
